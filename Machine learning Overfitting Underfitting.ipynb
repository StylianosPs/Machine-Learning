**----------------------------- F1 --------------------------------------**

```python
from turtle import color
import numpy as np
import math
import matplotlib.pyplot as plt

def dfx1(x1):
  return (2*(x1-2))

def dfx2(x2):
  return (2*(x2-3))

current_x1 = 0
current_x2 = 0
learning_rate = 0.5
T = 10
arrayFx = []
arrayX = [i for i in range(T)] 

for i in range(T):

  current_x1 = current_x1-learning_rate*dfx1(current_x1)
  current_x2 = current_x2-learning_rate*dfx2(current_x2)
  
  fx = pow(current_x1-2, 2) + pow(current_x2-3, 2)

  arrayFx.append(fx)
  
  print("Iteration: ",i,"\t(x1, x2) = ","(",current_x1,", ",current_x2,")","\tf1(x1,y1) = ",fx) 

plt.plot(arrayX, arrayFx, color = 'red')

plt.xlabel('iterations')
plt.ylabel('F(x1,x2)')
plt.show()
```

**Για τα default νούμερα παρατηρώ overflow στις πρώτες 3 επαναλήψεις **

```python
from turtle import color
import numpy as np
import math
import matplotlib.pyplot as plt

def dfx1(x1, x2):
  return (40*((x1+3)-pow((x2-3), 2)))

def dfx2(x1, x2):
  return (-2*(-x2+4)+40*(x1+3-pow(x2-3, 2))*(-2*x2+6))

current_x1 = 0
current_x2 = 0
learning_rate = 0.5
T = 10
arrayFx = []
arrayX = [i for i in range(T)] 

for i in range(T):
    prev_x1 = current_x1
    prev_x2 = current_x2

    current_x1 = current_x1-learning_rate*dfx1(prev_x1, prev_x2)
    current_x2 = current_x2-learning_rate*dfx2(prev_x1, prev_x2)
    
    fx = pow(1 - (current_x2 - 3), 2) + 20 * pow((current_x1 + 3) - pow((current_x2 - 3),2), 2)


    arrayFx.append(fx)
    
    print("Iteration: ",i,"\t(x1, x2) = ","(",current_x1,", ",current_x2,")","\tf1(x1,y1) = ",fx)  

plt.plot(arrayX, arrayFx, color = 'red')

plt.xlabel('iterations')
plt.ylabel('F(x1,x2)')
plt.show()
```
